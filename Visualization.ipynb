## Calculating ratio to get Relative Frequency.

```python
counts=fuel_econ['make'].value_counts()
print(counts)
print()
order2=counts.index
print(order2)
```

```python
plt.figure(figsize=[15,15])
sb.countplot(data=fuel_econ,y='make',order=order2)
for i,count in enumerate(counts):
    pct_string=str(round(float(100*(count/fuel_econ.shape[0])),4))+" %"
    plt.text(count/2,i,pct_string,va='center')
```

# Clustring.

#### Using Label Encoder.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
fuel_econ.id = le.fit_transform(fuel_econ.id)
fuel_econ.make = le.fit_transform(fuel_econ.make)
fuel_econ.model = le.fit_transform(fuel_econ.model)
fuel_econ.year = le.fit_transform(fuel_econ.year)
fuel_econ.VClass = le.fit_transform(fuel_econ.VClass)
fuel_econ.drive = le.fit_transform(fuel_econ.drive)
fuel_econ.trans =le.fit_transform(fuel_econ.trans)
fuel_econ.fuelType = le.fit_transform(fuel_econ.fuelType)
fuel_econ.cylinders =le.fit_transform(fuel_econ.cylinders)
fuel_econ.displ = le.fit_transform(fuel_econ.displ)
fuel_econ.pv2 = le.fit_transform(fuel_econ.pv2)
fuel_econ.pv4 = le.fit_transform(fuel_econ.pv4)
fuel_econ.city = le.fit_transform(fuel_econ.city)
fuel_econ.UCity = le.fit_transform(fuel_econ.UCity)
fuel_econ.highway = le.fit_transform(fuel_econ.highway)
fuel_econ.UHighway = le.fit_transform(fuel_econ.UHighway)
fuel_econ.comb = le.fit_transform(fuel_econ.comb)
fuel_econ.co2 = le.fit_transform(fuel_econ.co2)
fuel_econ.feScore = le.fit_transform(fuel_econ.feScore)
fuel_econ.ghgScore = le.fit_transform(fuel_econ.ghgScore)
fuel_econ.head()
```

#### Centroids.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3).fit(fuel_econ)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(fuel_econ['comb'] ,fuel_econ['displ'] , c= kmeans.labels_.astype(float), s=30, alpha=10)
plt.scatter(centroids[:, 0], centroids[:, 1],marker="x", c='black', s=100)
plt.show()
```

```python
plt.scatter(fuel_econ['UHighway'] ,fuel_econ['co2'] , c= kmeans.labels_.astype(float), s=30, alpha=10)
plt.scatter(centroids[:, 0], centroids[:, 1],marker="x", c='black', s=100)
plt.show()
```

```python
plt.scatter(fuel_econ['city'] ,fuel_econ['highway'] , c= kmeans.labels_.astype(float), s=30, alpha=10)
plt.scatter(centroids[:, 0], centroids[:, 1],marker="x", c='black', s=100)
plt.show()
```
